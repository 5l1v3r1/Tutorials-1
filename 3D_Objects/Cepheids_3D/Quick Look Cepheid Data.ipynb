```python
from astropy.table import Table
```

These data were taken from VizieR: ['Calibration of Cepheid Period/Luminosity relation (Fouque+, 2007)'](http://vizier.u-strasbg.fr/viz-bin/VizieR?-source=J/A%2BA/476/73). They are a merger of Tables 2 and 6 for all objects in Table 6. Merge performed in TopCat on "Star (Name)"; a subset of the merged columns were output to a FITS file.

```python
data = Table.read("Fouque2007_Cepheids.fits")
```

```python
data
```

```python
fig, ax = plt.subplots()
ax.scatter(data['_RAJ2000'], data['_DEJ2000'], c=data['E(B-V)'], s=100.*data['plx'], alpha=0.5)
ax.set_xlabel('RA', fontsize=20)
ax.set_ylabel('Dec', fontsize=20)
ax.set_title('Fouque2007_Cepheids')
ax.xaxis_inverted()
ax.grid(True)
fig.tight_layout()

plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(data['_RAJ2000'], data['Distance'], c=data['E(B-V)'], s=1000.*data['e_plx']/data['plx'], alpha=0.5)
ax.set_xlabel('RA', fontsize=20)
ax.set_ylabel('Distance (pc)', fontsize=20)
ax.set_title('Fouque2007_Cepheids')
ax.xaxis_inverted()
ax.grid(True)
fig.tight_layout()

plt.show()
```
